**Pokemon Data Set**

Import Libraries

```python
import requests
import json
URL="https://bulbapedia.bulbagarden.net/wiki/List_of_Pok%C3%A9mon_by_National_Pok%C3%A9dex_number#List_of_Pok.C3.A9mon_by_National_Pok.C3.A9dex_number"
page=requests.get(URL)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(page.content, 'html.parser')
```

```python
pokemon_content=soup.find(id='mw-content-text')

pokemon_tables=pokemon_content.find_all('table')
```

Getting Kanto Pokedex (KDex) location, as well as JohtoDex (JDex), etc.


```python
gen1_list=pokemon_tables[1]
gen1_list.contents
gen1_list.contents[1]
print(gen1_list.contents[1].contents[1].text.strip())

gen2_list=pokemon_tables[2]

print(gen2_list.contents[1].contents[1].text.strip())

gen3_list=pokemon_tables[8]

print(gen3_list.contents[1].contents[1].text.strip())
```

Finding which index to start at first

```python
gen1_list.contents[3]
```

Start at 3 and skip 1

Getting the first 3 pokemon of each generation/dex

```python
urlpoke='https://bulbapedia.bulbagarden.net/'
for j in range (1, 9):
    currentgen= pokemon_tables[j]
    info_start = 3
    dextype= currentgen.contents[1].contents[1].text.strip()
    gen = 'Generation ' + str(j) 
    print(f'{gen} {dextype}')
    for i in range(info_start, 8, 2):
        pokemon_info=currentgen.contents[i]
        kdex=pokemon_info.contents[1].text.strip()
        ndex=pokemon_info.contents[3].text.strip()
        name=pokemon_info.contents[7].text.strip()
        pokeurl= urlpoke+pokemon_info.contents[7].find("a")["href"]
        type1=pokemon_info.contents[9].text.strip()
        if len(pokemon_info.contents) > 10:
            type2=pokemon_info.contents[11].text.strip()
            print(f'Pokemon {kdex} {ndex} {name} find more at {pokeurl} is a {type1} & {type2} Pokemon')
        else:
            print(f'Pokemon {kdex} {ndex} {name} find more at {pokeurl} is a {type1} Pokemon')
```

Same loop but for every pokemon

```python
for j in range (1, 9):
    currentgen= pokemon_tables[j]
    info_start = 3
    dextype= currentgen.contents[1].contents[1].text.strip()
    gen_json_name = 'Pokemon ' + str(j) + ' gen.json'
    gen_json = []
    for i in range(info_start,  len(currentgen.contents), 2):
        pokemon_info=currentgen.contents[i]
        dex=pokemon_info.contents[1].text.strip()
        ndex=pokemon_info.contents[3].text.strip()
        name=pokemon_info.contents[7].text.strip()
        type1=pokemon_info.contents[9].text.strip()
        if len(pokemon_info.contents) > 10:
            type2=pokemon_info.contents[11].text.strip()
            gen_json.append({
                dextype: dex,
                "ndex": ndex,
                "name": name,
                "type1": type1,
                "type2": type2
            })
        else:
            gen_json.append({
                dextype: dex,
                "ndex": ndex,
                "name": name,
                "type1": type1
            })
    gen_json
    with open(gen_json_name, 'w') as outfile:
        json.dump(gen_json, outfile)
```
